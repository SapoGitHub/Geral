Sobre qubits, as matrizes de Pauli são as matrizes mais importantes:
\begin{equation}
\sigma_{x}=\left(\begin{array}{cc}
0 & 1\\
1 & 0
\end{array}\right),\quad\sigma_{y}=\left(\begin{array}{cc}
0 & -i\\
i & 0
\end{array}\right),\quad\sigma_{z}=\left(\begin{array}{cc}
1 & 0\\
0 & -1
\end{array}\right)
\end{equation}

Elas são:
* Hermitianas: $\sigma_{i}^{\dagger}=\sigma_{i}$
* Unitárias: $\sigma_{i}^{2}=1$

Outro conjunto de operadores são os 'abaixando' e 'levantando', definidos respectivamente como:
\begin{equation}
\sigma_{+}=\left|0\right\rangle \left\langle 1\right|=\left[\begin{array}{c}
1\\
0
\end{array}\right]\left[\begin{array}{cc}
0 & 1\end{array}\right]=\left[\begin{array}{cc}
0 & 1\\
0 & 0
\end{array}\right]
\end{equation}

e

\begin{equation}
\sigma_{-}=\left|1\right\rangle \left\langle 0\right|=\left[\begin{array}{c}
0\\
1
\end{array}\right]\left[\begin{array}{cc}
1 & 0\end{array}\right]=\left[\begin{array}{cc}
0 & 0\\
1 & 0
\end{array}\right]
\end{equation}

Eles estão relacionadas com as matrizes de Pauli como:
* $ \sigma_{x}=\sigma_{+}+\sigma_{-}=\left[\begin{array}{cc}
0 & 1\\
0 & 0
\end{array}\right]+\left[\begin{array}{cc}
0 & 0\\
1 & 0
\end{array}\right]=\left[\begin{array}{cc}
0 & 1\\
1 & 0
\end{array}\right] $
* $ \sigma_{y}=-i\left(\sigma_{+}-\sigma_{-}\right)=\left[\begin{array}{cc}
0 & -i\\
0 & 0
\end{array}\right]-\left[\begin{array}{cc}
0 & 0\\
-i & 0
\end{array}\right]=\left[\begin{array}{cc}
0 & -i\\
i & 0
\end{array}\right] $

Ou ainda:

\begin{equation}
\sigma_{\pm}=\frac{\sigma_{x}\pm i\sigma_{y}}{2}
\end{equation}

Esse nome vem da definição dos spins como $\left|0\right\rangle =\left|\uparrow\right\rangle$  e $\left|1\right\rangle =\left|\downarrow\right\rangle$, pois por exemplo:

\begin{equation}
\sigma_{+}\left|\downarrow\right\rangle =\left[\begin{array}{cc}
0 & 1\\
0 & 0
\end{array}\right]\left[\begin{array}{c}
0\\
1
\end{array}\right]=\left[\begin{array}{c}
1\\
0
\end{array}\right]=\left|\uparrow\right\rangle 
\end{equation}

Como calculamos anteriormente, podemos escrever a equação de estado de um bit como:

\begin{equation}
\left|\Psi\right\rangle =\cos\left(\frac{\theta}{2}\right)\left|0\right\rangle +e^{i\varphi}\sin\left(\frac{\theta}{2}\right)\left|1\right\rangle 
\end{equation}

Então podemos escrever na forma vetorial como:
\begin{equation}
\left|\Psi\right\rangle =\left[\begin{array}{cc}
\cos\left(\frac{\theta}{2}\right) & e^{i\varphi}\sin\left(\frac{\theta}{2}\right)\end{array}\right]^{T}
\end{equation}

E nossa matriz densidade vai ser então:
\begin{equation}
\rho=\left|\Psi\right\rangle \left\langle \Psi\right|=\left[\begin{array}{c}
\cos\left(\frac{\theta}{2}\right)\\
e^{i\varphi}\sin\left(\frac{\theta}{2}\right)
\end{array}\right]\left[\begin{array}{cc}
\cos\left(\frac{\theta}{2}\right) & e^{-i\varphi}\sin\left(\frac{\theta}{2}\right)\end{array}\right]
\end{equation}

\begin{equation}
\rho=\left[\begin{array}{cc}
\cos\left(\frac{\theta}{2}\right)^{2} & e^{-i\varphi}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right)\\
e^{i\varphi}\sin\left(\frac{\theta}{2}\right)\cos\left(\frac{\theta}{2}\right) & \sin\left(\frac{\theta}{2}\right)^{2}
\end{array}\right]
\end{equation}

Usando propriedades trigonométricas, então:
\begin{equation}
\rho=\left[\begin{array}{cc}
1+\cos\theta & \cos\varphi\sin\theta-i\sin\varphi\sin\theta\\
\cos\varphi\sin\theta+i\sin\varphi\sin\theta & 1-\cos\theta
\end{array}\right]
\end{equation}

E essa matriz pode ser decomposta em termos das matrizes de Pauli como:
\begin{equation}
\rho=\frac{1}{2}\left(I+\cos\varphi\sin\theta X+\sin\varphi\sin\theta Y+\cos\theta Z\right)
\end{equation}


\begin{equation}
\rho=\frac{1}{2}\left(I+\overrightarrow{r}_{b}\cdot\overrightarrow{\sigma}\right)
\end{equation}

Onde $\overrightarrow{\sigma}$ é um vetor das matrizes de Bloch $\overrightarrow{\sigma}=\left(\begin{array}{ccc}
\sigma_{x}, & \sigma_{y}, & \sigma_{z}\end{array}\right)$ e $\overrightarrow{r}_{b}$ é o vetor de Bloch $\overrightarrow{r}_{b}=\left(\begin{array}{ccc}
\cos\varphi\sin\theta, & \sin\varphi\sin\theta, & \cos\theta\end{array}\right)$.

Evidentemente, os operadores de rotação também podem ser expandidos nas matrizes de Pauli:
* $ R_{x}\left(\theta\right)=\cos\left(\frac{\theta}{2}\right)I-i\sin\left(\frac{\theta}{2}\right)X $
* $ R_{Y}\left(\theta\right)=\cos\left(\frac{\theta}{2}\right)I-i\sin\left(\frac{\theta}{2}\right)Y $
* $ R_{Z}\left(\theta\right)=\cos\left(\frac{\theta}{2}\right)I-i\sin\left(\frac{\theta}{2}\right)Z $

Isso é da teoria da física quântica, vem do operador momento angular no eixo z.

Por fim ainda temos que o raio do vetor de Bloch, se o ponto está na superfície indica um estado puro, quanto mais no interior, é um estado misto.

A matriz de densidade de um estado misto, é uma mistrua de prazes de estado puros:

\begin{equation}
\rho=\sum_{i=1}^{k}c_{i}\rho_{i}
\end{equation}

Onde $0\leq c_{i}\leq1$ e $\sum_{i=1}^{k}c_{i}=1$. Então lembrando da definição de uma matriz de estado pura:

\begin{equation}
\rho_{i}=\frac{1}{2}\left(I+\overrightarrow{r}_{i}\cdot\overrightarrow{\sigma}\right)
\end{equation}

Então:

\begin{equation}
\rho=\sum_{i=1}^{k}c_{i}\frac{1}{2}\left(I+\overrightarrow{r}_{i}\cdot\overrightarrow{\sigma}\right)
\end{equation}

\begin{equation}
\rho=\frac{1}{2}\left(I+\sum_{i=1}^{k}\left(c_{i}\overrightarrow{r}_{i}\right)\cdot\overrightarrow{\sigma}\right)
\end{equation}

\begin{equation}
\rho=\frac{1}{2}\left(I+\overrightarrow{r}_{m}\cdot\overrightarrow{\sigma}\right)
\end{equation}

Então agora o raio do nosso estado misturado é dado por
\begin{equation}
\overrightarrow{r}_{m}=\sum_{i=1}^{k}\left(c_{i}\overrightarrow{r}_{i}\right)
\end{equation}

Por exemplo, se nosso bit ao invés de estar em $\left|0\right\rangle $ __e__ $\left|1\right\rangle$ está em $\left|0\right\rangle $ __ou__ $\left|1\right\rangle $, com a mesma probabilidade. Isto é, um estado possível é dado por:
\begin{equation}
\left|\psi_{0}\right\rangle =\left|0\right\rangle 
\end{equation}
E outro estado é:
\begin{equation}
\left|\psi_{1}\right\rangle =\left|1\right\rangle 
\end{equation}
Então as matrizes de densidade são:
\begin{equation}
\rho_{0}=\left|0\right\rangle \left\langle 0\right|=\left[\begin{array}{c}
1\\
0
\end{array}\right]\left[\begin{array}{cc}
1 & 0\end{array}\right]=\left[\begin{array}{cc}
1 & 0\\
0 & 0
\end{array}\right]
\end{equation}

E

\begin{equation}
\rho_{1}=\left|1\right\rangle \left\langle 1\right|=\left[\begin{array}{c}
0\\
1
\end{array}\right]\left[\begin{array}{cc}
0 & 1\end{array}\right]=\left[\begin{array}{cc}
0 & 0\\
0 & 1
\end{array}\right]
\end{equation}

Então a matriz de densidade vai ser dada por:
\begin{equation}
\rho=p_{0}\left|0\right\rangle \left\langle 0\right|+p_{1}\left|1\right\rangle \left\langle 1\right|
\end{equation}

Como vamos considerar que tem 50% de probabilidade de estar em cada estado:
\begin{equation}
\rho=\frac{1}{2}\left|0\right\rangle \left\langle 0\right|+\frac{1}{2}\left|1\right\rangle \left\langle 1\right|=\left[\begin{array}{cc}
\frac{1}{2} & 0\\
0 & \frac{1}{2}
\end{array}\right]
\end{equation}

Decompondo cada um dos estados nas matrizes de Pauli, ficamos com:
\begin{equation}
\rho_{0}=\frac{1}{2}\left(I+\left(0,0,1\right)\cdot\overrightarrow{\sigma}\right)
\end{equation}
E:
\begin{equation}
\rho_{1}=\frac{1}{2}\left(I+\left(0,0,-1\right)\cdot\overrightarrow{\sigma}\right)
\end{equation}

Então o raio de Bloch do nosso estado misto vai ser:
\begin{equation}
\overrightarrow{r}_{m}=\sum_{i=0}^{1}\left(c_{i}\overrightarrow{r}_{i}\right)=\frac{1}{2}\left(0,0,1\right)+\frac{1}{2}\left(0,0,-1\right)=\left(0,0,0\right)
\end{equation}

```python
import numpy as np

# Nossa matriz de estado:
p=[0]*2
p[0]=np.array([[1,0],[0,0]])
p[1]=np.array([[0,0],[0,1]])


# Definimos as matrizes de pauli:
pauli = {
    "I": np.array([[1,0],[0,1]]),
    "X": np.array([[0,1],[1,0]]),
    "Y": np.array([[0,-1j],[1j,0]]),
    "Z": np.array([[1,0],[0,-1]])
}

# Agora vamos calcular o valor dos elementos efetivamente:
coeficientes={}
for b in pauli:
    rotulo = b
    produto = p[0].dot(pauli[b])   # Fazemos o produto entre as matrizes
    coeficientes[rotulo]=produto.trace()
    
coeficientes                    # Exibimos os coeficientes
```
