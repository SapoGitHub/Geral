# Introdução

## 1.1 Representação de dados e similaridade

### 1.1.1 Dados de treinamento

* Problema: temos duas classes de objetos, então se nos depararmos com um novo objeto, preisamos classificá-lo em uma das classes.

Dados empírcos:

\begin{equation}
\left(x_{1},y_{1}\right),\dots,\left(x_{m},y_{m}\right)\in\mathcal{X}\times\left\{ \pm1\right\} 
\end{equation}

* $ \mathcal{X} $: conjunto de onde retiramos os padrões $x_{i}$, chamado também de domínio:
  * $ x_{i}\rightarrow $ entradas, casos, instâncias, observações.
* $\left\{ \pm1\right\}$: onde retiramos $y_{i}$:
  * $y_{i}\rightarrow $ rótulos, alvos, saídas, observações;
  * $\left\{ \pm1\right\} \rightarrow$ somente duas classes, padrão de reconhecimento (binário) ou classificação (binária).

Precisamos de um tipo adicional de estrutura, para generalizar os dados que não estamos vendo:

* Dado um novo padrão $ x\in\mathcal{X}$, queremos predizer o correspondente $y\in\left\{ \pm1\right\}$ ;
 * Fazendo isso para cada $x\in\mathcal{X}$ nos leva a estimar uma função $f\rightarrow X\rightarrow\left\{ \pm1\right\}$.
* Queremos escolher y de forma que $\left(x,y\right)$ seja coerente aos exemplos de treinamento.
* Precisamos ter noções de similaridade em $\mathcal{X}$ e $\left\{ \pm1\right\}$.

A similaridade da saída é fácil:
* Temos somente duas situações, os rótulos podem ser: idênticos ou diferente.

A similaridade das entradas é complicado e constitui o núcleo do aprendizado de máquina. Considerando uma medida de similaridade na forma:

\begin{equation}
k:\mathcal{X}\times\mathcal{X}\rightarrow\mathbb{R}
\end{equation}

\begin{equation}
\left(x,x\right)\rightarrow k\left(x,x'\right)
\end{equation}

* Dados dois padrões x e x', retorna um número real, caracterizando sua similaridade.
* Assumimos que k é simétrico, isto é: $k\left(x,x'\right)=k\left(x',x\right)$ para todo $x,x'\in\mathcal{X}$.
* $k\rightarrow$ é chamado de kernel.

### Produto interno

Medidas gerais de similaridade são difíceis de serem estudadas. Então começaremos com um caso particular, um tipo simples de similaridade é o produto interno.

Para dois vetores $\overrightarrow{x},\overrightarrow{x'}\in\mathbb{R}^{\mathcal{N}}$, o produto interno canônico é definido como:

\begin{equation}
\left(x,x'\right)=\sum_{i=1}^{\mathcal{N}}\left[x\right]_{i}\left[x'\right]_{i}
\end{equation}

* $\left[x\right]_{i}$ é o elemento i de $\overrightarrow{x}$.

### Distância

A interpretação geométrica do produto interno canônico é que ele computa o ângulo entre os vetores $\overrightarrow{x}$ e $\overrightarrow{x'}$, desde que sejam normalizados. Ele também permite o cálculo da norma de um vetor x como:

\begin{equation}
\left\Vert \overrightarrow{x}\right\Vert =\sqrt{\left\langle x,x\right\rangle }
\end{equation}


* A distância entre dois vetores, é computado como o comprimento do vetor diferença.
* Então calcular produto interno significa ser capaz de lidar com todas construções geométricas.

Observação: produto interno não é suficientemente geral para lidar com todos problemas.

* Precisamos ser capaz de representar as entradas como vetores em algum espaço $\mathcal{H}$ (que não precisa ser $\mathbb{R}^{\mathcal{N}}$). Então usamos o mapa:

\begin{equation}
\Phi=\mathcal{X}\rightarrow\mathcal{H}
\end{equation}

\begin{equation}
x\rightarrow\overrightarrow{x}=\Phi\left(x\right)
\end{equation}

* Mesmo que os padrões originais existam em um _dot product space_, podemos desejar medidas de similaridade mais gerais obtidas com a aplicação de um mapa, nesse caso $\Phi$ normalmente será um mapa não linear.
  * _Dot product space (inner product space)_ : espaço em que há produto interno.

### Feature Space

* $\mathcal{H}\rightarrow$ _feature space_.
* $\overrightarrow{x}\rightarrow$ denota a representação vetorial de x no _feature space_.
* A medida de similaridade do produto interno em $\mathcal{H}$ é:

\begin{equation}
k\left(x,x'\right)=\left\langle \overrightarrow{x}.\overrightarrow{x'}\right\rangle =\left\langle \Phi\left(x\right).\Phi\left(x'\right)\right\rangle 
\end{equation}

* Podemos então lidar com padrões geométricos e estudar algoritmos de aprendizado usando álgebra linear e geometria análica (que pode resolver problemas não-lineares).

### Resumo
Pelo que entendi então:

* Vamos ter um conjunto de entradas $\mathcal{X}=\left\{ x_{i}\right\}$.
* Queremos classificar em $\mathcal{Y}=\left\{ \pm1\right\}$;
* Usamos uma função (mapa) $\Phi\left(x\right)$ para transformar as entradas em um vetor $x\rightarrow\overrightarrow{x}$, levando para um espaço de dimensão maior ($\mathcal{H}$, _feature space_);
* Neste espaço usamos um kernel pra medir as distâncias entre dois vetores , que o caso simples é o produto interno: $ k=\left\langle x_{i}|x_{j}\right\rangle $
* Podemos supor que isso  gera uma matriz simétrica onde os elementos são $M_{ij}=\left\langle x_{i}|x_{j}\right\rangle$.

## Algoritmo simples de reconhecimento de padrões

* Vamos descrever o algorimo mais simples possível para reconhecimento de padrões.
* Assumimos que o dado está embutido em um espaço com protudo interno $\mathcal{H}^{3}$.
  * Com produto interno, podemos medir distâncias neste espaço.
* Objetivo: associar uma entrada não vista no treinamento, a classe com a média mais próxima.

Vamos começar calculando a média das duas classes no espaço de destaque:

* $c_{+}=\frac{1}{m_{+}}\sum_{\left\{ i|y_{i}=+1\right\} }\overrightarrow{x_{i}}$
* $c_{-}=\frac{1}{m_{-}}\sum_{\left\{ i|y_{i}=-1\right\} }\overrightarrow{x_{i}}$

Onde:
• $m_{\pm}$ são o número de exemplos com rótulos $\pm$.
• Nós vamos associar associamos ao ponto $\overrightarrow{x}$ a classe com a média mais próxima.
• Isso pode ser formulado como produto interno.

O ponto médio entre $\overrightarrow{c}_{+}$ e $\overrightarrow{c}_{-}$é:

\begin{equation}
\overrightarrow{c}=\frac{\overrightarrow{c}_{+}+\overrightarrow{c}_{-}}{2}
\end{equation}

Definimos a classe de $\overrightarrow{x}$ checando se o vetor $\overrightarrow{a}=\overrightarrow{x}-\overrightarrow{c}$ tem um ângulo menor que $\frac{\pi}{2}$ com o vetor $\overrightarrow{w}=\overrightarrow{c}_{+}-\overrightarrow{c}_{-} $.

* $\overrightarrow{a}=\overrightarrow{x}-\overrightarrow{c}\rightarrow$ distância entre $\overrightarrow{x} e \overrightarrow{c}$
* $\overrightarrow{w}=\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\rightarrow$ distância entre as médias dos rótulos de saída.
* Então se $\overrightarrow{a}$ tem menos de $\frac{\pi}{2}$ com $\overrightarrow{w}$,ele está do lado de $\overrightarrow{C}$ mais próximo a $C_{+1}$.

Isso leva a:

\begin{equation}
y=\text{sgn}\left\langle \overrightarrow{a},\overrightarrow{w}\right\rangle 
\end{equation}

\begin{equation}
y=\text{sgn}\left\langle \left(\overrightarrow{x}-\overrightarrow{c}\right),\left(\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\right)\right\rangle 
\end{equation}

\begin{equation}
y=\text{sgn}\left\langle \overrightarrow{x}-\left(\frac{\overrightarrow{c}_{+}+\overrightarrow{c}_{-}}{2}\right),\left(\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\right)\right\rangle 
\end{equation}


Manipulando o produto interno, pra números reais, e vamos trabalhar apenas com o eixo x, pois os outros eixos seriam análogos:

\begin{equation}
\left[\left\langle \overrightarrow{x}-\left(\frac{\overrightarrow{c}_{+}+\overrightarrow{c}_{-}}{2}\right),\left(\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\right)\right\rangle \right]_{x}=\left(\overrightarrow{x}-\left(\frac{\overrightarrow{c}_{+}+\overrightarrow{c}_{-}}{2}\right)\right)_{x}\left(\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\right)_{x}
\end{equation}

\begin{equation}
=\left(\frac{2x_{x}-c_{+x}-c_{-x}}{2}\right)\left(c_{+x}-c_{-x}\right)+
\end{equation}

\begin{equation}
=\frac{1}{2}\left[\left(2x_{x}-c_{+x}-c_{-x}\right)\left(c_{+x}-c_{-x}\right)\right]
\end{equation}

\begin{equation}
=\frac{1}{2}\left[\left(2x_{x}c_{+x}-c_{+x}c_{+x}-c_{-x}c_{+x}\right)-\left(2x_{x}c_{-x}-c_{+x}c_{-x}-c_{-x}c_{-x}\right)\right]
\end{equation}

\begin{equation}
=\frac{1}{2}\left[\left(2x_{x}c_{+x}-c_{+x}^{2}-c_{-x}c_{+x}\right)-\left(2x_{x}c_{-x}-c_{+x}c_{-x}-c_{-x}^{2}\right)\right]
\end{equation}

\begin{equation}
=\frac{1}{2}\left[\left(2x_{x}c_{+x}-2x_{x}c_{-x}\right)-c_{+x}^{2}+c_{-x}^{2}-c_{-x}c_{+x}+c_{+x}c_{-x}\right]
\end{equation}

\begin{equation}
=\left(x_{x}c_{+x}-x_{x}c_{-x}\right)+\frac{1}{2}\left(c_{-x}^{2}-c_{+x}^{2}\right)
\end{equation}


Generalizando:
\begin{equation}
\left\langle \overrightarrow{x}-\left(\frac{\overrightarrow{c}_{+}+\overrightarrow{c}_{-}}{2}\right),\left(\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\right)\right\rangle =\left(x_{x}c_{+x}+x_{y}c_{+y}\right)-\left(x_{x}c_{-x}+x_{y}c_{-y}\right)+\frac{1}{2}\left[\left(c_{-x}^{2}+c_{-y}^{2}\right)-\left(c_{+x}^{2}+c_{+y}^{2}\right)\right]
\end{equation}

Pelas definições de produto interno então:

\begin{equation}
\left\langle \overrightarrow{x}-\left(\frac{\overrightarrow{c}_{+}+\overrightarrow{c}_{-}}{2}\right),\left(\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\right)\right\rangle =\left\langle \overrightarrow{x}|\overrightarrow{c}_{+}\right\rangle -\left\langle \overrightarrow{x}|\overrightarrow{c}_{-}\right\rangle +\frac{1}{2}\left[\left(c_{-x}^{2}+c_{-y}^{2}\right)-\left(c_{+x}^{2}+c_{+y}^{2}\right)\right]
\end{equation}


E chamando o último termo de:

\begin{equation}
b=\frac{1}{2}\left[\left(c_{-x}^{2}+c_{-y}^{2}\right)-\left(c_{+x}^{2}+c_{+y}^{2}\right)\right]
\end{equation}

Lembrando que:

\begin{equation}
\left\Vert c_{-}\right\Vert =\sqrt{\left\langle c_{-}|c_{-}\right\rangle }=\sqrt{\left(c_{-x}^{2}+c_{-y}^{2}\right)}
\end{equation}

Então, chamamos isso de offset:

\begin{equation}
b=\frac{1}{2}\left(\left\Vert c_{-}\right\Vert ^{2}-\left\Vert c_{+}\right\Vert ^{2}\right)
\end{equation}

\begin{equation}
b=\frac{1}{2}\left(\left\langle c_{-}|c_{-}\right\rangle -\left\langle c_{+}|c_{+}\right\rangle \right)
\end{equation}

* Uma medida da diferença de tamanho entre as médias.

E ficamos com:

\begin{equation}
\left\langle \overrightarrow{x}-\left(\frac{\overrightarrow{c}_{+}+\overrightarrow{c}_{-}}{2}\right),\left(\overrightarrow{c}_{+}-\overrightarrow{c}_{-}\right)\right\rangle =\left\langle \overrightarrow{x}|\overrightarrow{c}_{+}\right\rangle -\left\langle \overrightarrow{x}|\overrightarrow{c}_{-}\right\rangle +b
\end{equation}

Logo:

\begin{equation}
y=\text{sgn}\left(\left\langle \overrightarrow{x}|\overrightarrow{c}_{+}\right\rangle -\left\langle \overrightarrow{x}|\overrightarrow{c}_{-}\right\rangle +b\right)
\end{equation}


* A equação anterior induz um limite de decisão que tem a forma de um hiperplano.
  * Um conjunto de pontos que satisfazem a restrição expressa como uma equação linear.

Vamos reescrever em termos dos padrões de entrada $x_{i}$ e usando o kernel k para calcular os produtos internos. 

* Sabemos apenas como calcular o produto interno entre representações vetoriais de $\overrightarrow{x}_{i}$ de $x_{i}$, precisamos então expressar os vetores $\overrightarrow{c}_{i}$ e $\overrightarrow{w}$ em termos de $\overrightarrow{x}_{i}$.

Então:
\begin{equation}
y=\text{sgn}\left(\left\langle \overrightarrow{x}|\overrightarrow{c}_{+}\right\rangle -\left\langle \overrightarrow{x}|\overrightarrow{c}_{-}\right\rangle +b\right)
\end{equation}

\begin{equation}
y=\text{sgn}\left(\left\langle \overrightarrow{x}|\frac{1}{m_{+}}\sum_{\left\{ i|y_{i}=+1\right\} }\overrightarrow{x_{i}}\right\rangle -\left\langle \overrightarrow{x}|\frac{1}{m_{-}}\sum_{\left\{ i|y_{i}=-1\right\} }\overrightarrow{x_{i}}\right\rangle +b\right)
\end{equation}

\begin{equation}
y=\text{sgn}\left(\frac{1}{m_{+}}\sum_{\left\{ i|y_{i}=+1\right\} }\left\langle \overrightarrow{x}|\overrightarrow{x_{i}}\right\rangle -\frac{1}{m_{-}}\sum_{\left\{ i|y_{i}=-1\right\} }\left\langle \overrightarrow{x}|\overrightarrow{x_{i}}\right\rangle +b\right)
\end{equation}

\begin{equation}
y=\text{sgn}\left(\frac{1}{m_{+}}\sum_{\left\{ i|y_{i}=+1\right\} }k\left(x,x_{i}\right)-\frac{1}{m_{-}}\sum_{\left\{ i|y_{i}=-1\right\} }k\left(x,x_{i}\right)+b\right)
\end{equation}


E o deslocamento assume:

\begin{equation}
b=\frac{1}{2}\left(\left\langle c_{-}|c_{-}\right\rangle -\left\langle c_{+}|c_{+}\right\rangle \right)
\end{equation}

\begin{equation}
b=\frac{1}{2}\left(\left\langle \frac{1}{m_{-}}\sum_{\left\{ i|y_{i}=-1\right\} }\overrightarrow{x_{i}}|\frac{1}{m_{-}}\sum_{\left\{ j|y_{j}=-1\right\} }\overrightarrow{x_{j}}\right\rangle -\left\langle \frac{1}{m_{+}}\sum_{\left\{ i|y_{i}=+1\right\} }\overrightarrow{x_{i}}|\frac{1}{m_{+}}\sum_{\left\{ j|y_{j}=+1\right\} }\overrightarrow{x_{j}}\right\rangle \right)
\end{equation}

\begin{equation}
b=\frac{1}{2}\left(\frac{1}{m_{-}^{2}}\sum_{\left\{ i,j|y_{i}=y_{j}=-1\right\} }\left\langle \overrightarrow{x_{i}}|\overrightarrow{x_{j}}\right\rangle -\frac{1}{m_{+}^{2}}\sum_{\left\{ i,j|y_{i}=y_{j}=+1\right\} }\left\langle \overrightarrow{x_{i}}|\overrightarrow{x_{j}}\right\rangle \right)
\end{equation}

\begin{equation}
b=\frac{1}{2}\left(\frac{1}{m_{-}^{2}}\sum_{\left\{ i,j|y_{i}=y_{j}=-1\right\} }k\left(x_{i},x_{j}\right)-\frac{1}{m_{+}^{2}}\sum_{\left\{ i,j|y_{i}=y_{j}=+1\right\} }k\left(x_{i},x_{j}\right)\right)
\end{equation}

### Meu exemplo 2D (Python)

* Vamos trabalhar apenas com a parte positiva do eixo y.
* Nossas classes vão ser o lado positivo e negativo do eixo x.

```python
import numpy as np                 # Biblioteca fom funções matemáticas
import matplotlib.pyplot as plt    # Biblioteca para geração de gráficos
```

Para treinamento, vamos gerar 100 pontos em cada lado do eixo.
* Para os valores positivos (vamos chamar assim os valores no lado positivo do eixo x), os valores que podem assumir são 0<x<100 e 0<y<100
* Para os valores negativos, temos de forma análoga -100<x<0 e 0<y<100.

```python
# Função para gerar dados de treinamento
def dados_trei (n):
    # n - Quantidade dados
    pos=[]   # Lista dos pontos positivos
    neg=[]   # Lista dos pontos negativos

    for i in range(n):  # Vamos gerar 100 pontos de cada classe   
        #Vamos gerar de forma aleatória
        pos.append(np.array([np.random.randint(low=0, high=100, size=1)[0], np.random.randint(low=1, high=100, size=1)[0]]) )
        neg.append(np.array([-np.random.randint(low=1, high=100, size=1)[0], np.random.randint(low=1, high=100, size=1)[0]]))
        
    return (pos,neg)
```

```python
# Função pra retornar duas listas com as componentes X e Y de uma lista de Arrays
def plotar (lista):
    # lista - Lista original de arrays
    
    n = len(lista)    # tamanho da lista
    x = []            # Lista com os valores do eixo X
    y = []            # Lista com os valores do eixo Y
    
    # Separamos os valores 
    for i in range (n):
        x.append(lista[i][0])
        y.append(lista[i][1])
        
    return (x,y)    # Retornamos a tupla
```

```python
# Vamos plotar os dados de treinamento

(pos,neg) = dados_trei(1000)  # Geramos os dados de treinamento
(px,py) = plotar (pos)       # Separamos os pontos positivox em X e Y
(nx,ny) = plotar (neg)       # Separaos os pontos negativos em X e Y
    

plt.plot(px, py,'ob', label='Positivo')
plt.plot(nx, ny,'or', label='Negativo')

# Configurando o plot
plt.title("Pontos de treinamento")
plt.legend()
plt.show()
```

```python
# Função para criar o vetor médio da classe
def med (p):
    # p - Array de pontos
    m = np.array([0,0])  # Média inicial
    n = len(p)           # Quantidade de pontos
    
    for i in range (n):
        m=m+p[i]
    m=m/n
    return m
```

```python
# Vamos calcular as médias das classes:
cpos = med(pos)
cneg = med(neg)

plt.plot(cpos[0],cpos[1],'ob', label='Positivo')
plt.plot(cneg[0],cneg[1],'or', label='Negativo')

# Configurando o plot
plt.title("Média das classes")
plt.legend()
plt.ylim(0, 100)    # Eixo Y
plt.xlim(-100, 100) # Eixo X
plt.show()
```

A partir daqui os dados de treinamento são irrelevantes. Então agora vamos classificar outros pontos.

```python
# Função pra gerar os dados de teste
def dados_tes(n):
    pts=[]

    for i in range(n):  
        #Vamos gerar de forma aleatória
        pts.append(np.array([np.random.randint(low=-100, high=100, size=1)[0], np.random.randint(low=1, high=100, size=1)[0]]) )
        
    return pts
```

```python
#Vamos gerar os dados de treinamento

pts = dados_tes(2000)

(ptx,pty) = plotar(pts)
    
plt.plot(ptx,pty,'o',color='black', label='Pontos')
plt.plot(cpos[0],cpos[1],'ob', label='Positivo',markersize=20)
plt.plot(cneg[0],cneg[1],'or', label='Negativo',markersize=20)

#Configurando o plot
plt.title("Pontos de treinamento")
plt.legend()
plt.show()
```

```python
# Agora classificamos

pos=[]  # Lista para guardar os pontos positivos
neg=[]  # Lista para guardar os pontos negativos

c = (cpos+cneg)/2 # O ponto médio entre as médias
w = cpos-cneg     # A distância entre as médias
mod_w = np.sqrt(w.dot(w)) # Módulo de w

for i in range(len(pts)):
    a = pts[i]-c                # A distância  enter o ponto e a média
    mod_a=np.sqrt(a.dot(a))     # Módulo de a
    pi = a.dot(w)/(mod_w*mod_a) # O produto interno entre a e w normalizado.
    ang = np.arccos(pi)         # Arccos do produto interno
    if (ang<(np.pi/2)):         # Se ele é menor, então classificamos como c+, positivo
        pos.append(pts[i])
    else:
        neg.append(pts[i])
    
```

```python
# Vamos plotar
(px,py) = plotar (pos)
(nx,ny) = plotar (neg)

    
plt.plot(px,py,'og', label='Pontos positivos')
plt.plot(nx,ny,'oy', label='Pontos negativos')

plt.plot(cpos[0],cpos[1],'ob', label='Positivo',markersize=20)
plt.plot(cneg[0],cneg[1],'or', label='Negativo',markersize=20)

plt.plot([0,0],[0,100],color='black')#'-b')

#Configurando o plot
plt.title("Pontos")
plt.legend()
plt.show()
```

### Meu exemplo (Direto)

Algumas coisas que estão evidentes na literatura e até em nossa formulação matemática final, mas ficou despercebido aqui, são:
* $\overrightarrow{w}=\overrightarrow{c}_{+}-\overrightarrow{c}_{-}$ aponta no sentido de $\overrightarrow{c}_{+}$
* $\overrightarrow{a}=\overrightarrow{x}_{+}-\overrightarrow{c}$ sai do ponto médio entre $\overrightarrow{c}_{-}$ e $\overrightarrow{c}_{+}$ na direção de $\overrightarrow{x}$.

Então se a projeção de $\overrightarrow{a}$ em $\overrightarrow{w}$ ($\left\langle \overrightarrow{a}|\overrightarrow{w}\right\rangle $) tiver sinal positivo, significa que aponta no mesmo sentido de $\overrightarrow{w}$, isto é, para $\overrightarrow{c}_{+}$. Por isso não precisamos computar o ângulo diretamente, e nem normalizar.

Vamos refazer nosso exemplo desta maneira.

```python
# Função pra refletir um conjunto de pontos
def ref (pt):
    r=[]  # Lista de Arrays refletidas
    
    # pt - Lista de Arrays de pontos a serem refletidos
    for i in range(500):
        x=-pos[i][0]
        y=pos[i][1]
        r.append(np.array([x,y]))
        
    return r
```

```python
# Etapa de treinamento
(pos,neg) = dados_trei(500)   # Vamos gerar os dados de treinamento

neg = ref(pos)

cpos = med(pos)               # Calcular o vetor do núcleo positivo
cneg = med(neg)               # Calcular o vetor do núcleo negativo
```

```python
# Etapa de execução

## Vamos calcular o valor de b

### Parte positiva
bpos = 0
for x  in range(len(pos)):
    for y in range(len(pos)):
        bpos=bpos+pos[x].dot(pos[y])
bpos=bpos/(len(pos)*len(pos))

### Parte negativa
bneg = 0
for x  in range(len(neg)):
    for y in range(len(neg)):
        bneg=bneg+neg[x].dot(neg[y])
bneg=bneg/(len(neg)*len(neg))

b=(bpos-bneg)/2
```
